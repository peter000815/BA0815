```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('AMZN (1).csv')
x = data['Date']
y = data['Close']
plt.figure(figsize=(18,4)) #创建绘图对象
plt.plot(x,y,"blue",linewidth=1) #在当前绘图对象绘图（X轴，Y轴，蓝色虚线，线宽度）
plt.xticks(x)                           #设置x轴上的刻度 
plt.xticks(x[::4])                    #当刻度太密集时候使用列表的步长（间隔取值）来解决
plt.xlabel("Dates")
plt.ylabel("Stock price($)") #Y轴标签
plt.title("Amazon stock") #图标题
plt.show() #显示图
```
